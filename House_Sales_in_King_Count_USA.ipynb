# House Sales in King County, USA

# About the Dataset

This dataset contains house sale prices for King County, which includes Seattle. It includes homes sold between May 2014 and May 2015. It was taken from [here](https://www.kaggle.com/harlfoxem/housesalesprediction?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-wwwcourseraorg-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDA0101ENSkillsNetwork20235326-2022-01-01). It was also slightly modified for the purposes of this course. 


| Variable      | Description                                                                                                 |
| ------------- | ----------------------------------------------------------------------------------------------------------- |
| id            | A notation for a house                                                                                      |
| date          | Date house was sold                                                                                         |
| price         | Price is prediction target                                                                                  |
| bedrooms      | Number of bedrooms                                                                                          |
| bathrooms     | Number of bathrooms                                                                                         |
| sqft_living   | Square footage of the home                                                                                  |
| sqft_lot      | Square footage of the lot                                                                                   |
| floors        | Total floors (levels) in house                                                                              |
| waterfront    | House which has a view to a waterfront                                                                      |
| view          | Has been viewed                                                                                             |
| condition     | How good the condition is overall                                                                           |
| grade         | overall grade given to the housing unit, based on King County grading system                                |
| sqft_above    | Square footage of house apart from basement                                                                 |
| sqft_basement | Square footage of the basement                                                                              |
| yr_built      | Built Year                                                                                                  |
| yr_renovated  | Year when house was renovated                                                                               |
| zipcode       | Zip code                                                                                                    |
| lat           | Latitude coordinate                                                                                         |
| long          | Longitude coordinate                                                                                        |
| sqft_living15 | Living room area in 2015(implies-- some renovations) This might or might not have affected the lotsize area |
| sqft_lot15    | LotSize area in 2015(implies-- some renovations)                                                            |


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr
```

# Data Wrangling


```python
df.drop(columns=['id','Unnamed: 0'], inplace=True)
df.describe()
```

```python
print("number of NaN values for the column bedrooms :", df['bedrooms'].isnull().sum())
print("number of NaN values for the column bathrooms :", df['bathrooms'].isnull().sum())
```

```python
mean=df['bedrooms'].mean()
df['bedrooms'].replace(np.nan,mean, inplace=True)
```

```python
mean=df['bathrooms'].mean()
df['bathrooms'].replace(np.nan,mean, inplace=True)
```

```python
print("number of NaN values for the column bedrooms :", df['bedrooms'].isnull().sum())
print("number of NaN values for the column bathrooms :", df['bathrooms'].isnull().sum())
```

# Model Development


### Single Regression Model

```python
X = df[['long']]
Y = df['price']
lm = LinearRegression()
lm.fit(X,Y)
s_yhat = lm.predict(X)
print('R^2 = ',lm.score(X, Y),'\nMSE = ',mean_squared_error(s_yhat, Y))
```

```python
sns.regplot(df,x='long', y='price', line_kws={'color':'red'})
```

```python
X = df[['sqft_living']]
Y = df['price']
lr = LinearRegression()
lr.fit(X,Y)
s_yhat = lr.predict(X)
print('R^2 = ',lr.score(X, Y),'\nMSE = ',mean_squared_error(s_yhat, Y))
```

### Multiple Regression Model

```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]     
```

```python
Z= df[features]  
z_lr = LinearRegression()
z_lr.fit(Z,Y)
m_yhat = lr.predict(X)
z_score = z_lr.score(Z,Y)
print('R^2 = ',z_score,'\nMSE = ',mean_squared_error(m_yhat, Y))
```

```python
ax1 = sns.distplot(Y, hist=False, color='b', label='Actual Values')
sns.distplot(m_yhat, hist=False, color='r',label='Predicted Values', ax=ax1)

plt.title('Actual vs Predicted Values')
plt.legend(['Actual Values','Predicted Values'])
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Houses')
plt.show()
```

### Polynomail Regression

```python
Input=[('scale',StandardScaler()),('polynomial', PolynomialFeatures(include_bias=False)),('model',LinearRegression())]
```

```python
pipe = Pipeline(Input)
pipe.fit(Z,Y)
p_yhat = pipe.predict(Z)
z_pipe_score = pipe.score(Z,Y)
print('R^2 = ',z_pipe_score,'\nMSE = ',mean_squared_error(p_yhat, Y))
```

```python
ax1 = sns.distplot(Y, hist=False, color='b', label='Actual Values')
sns.distplot(p_yhat, hist=False, color='r',label='Predicted Values', ax=ax1)

plt.title('Actual vs Predicted Values')
plt.legend(['Actual Values','Predicted Values'])
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Houses')
plt.show()
```

# Model Evaluation and Refinement


```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]    
X = df[features]
Y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)


print("number of test samples:", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
RR = Ridge(alpha=0.1)
RR.fit(x_train,y_train)
r_yhat = RR.predict(x_test)
ridge_score = RR.score(x_test, y_test)

print('R^2 = ',ridge_score,'\nMSE = ',mean_squared_error(r_yhat, y_test))
```

```python
ax1 = sns.distplot(Y, hist=False, color='b', label='Actual Values')
sns.distplot(r_yhat, hist=False, color='r',label='Predicted Values', ax=ax1)

plt.title('Actual vs Predicted Values using test data')
plt.legend(['Actual Values','Predicted Values'])
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Houses')
plt.show()
```

### Polynomial Ridge Regression

```python
pr = PolynomialFeatures()
x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.fit_transform(x_test)

RR_poly = Ridge(alpha=0.1)
RR_poly.fit(x_train_pr,y_train)
poly_yhat = RR_poly.predict(x_test_pr)
RR_poly_score = RR_poly.score(x_test_pr,y_test)

print('R^2 = ',RR_poly_score,'\nMSE = ',mean_squared_error(poly_yhat, y_test))
```

```python
ax1 = sns.distplot(Y, hist=False, color='b', label='Actual Values')
sns.distplot(poly_yhat, hist=False, color='r',label='Predicted Values', ax=ax1)

plt.title('Actual vs Predicted Values using test data')
plt.legend(['Actual Values','Predicted Values'])
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Houses')
plt.show()
```
